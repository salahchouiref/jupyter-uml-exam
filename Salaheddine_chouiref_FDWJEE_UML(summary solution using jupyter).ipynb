## Solution -> Exercice 1

### 1. Identification des Acteurs
- **Client** : Un individu possédant un compte bancaire et une carte de crédit (Visa ou carte de la banque).
- **Banque** : L'institution financière qui fournit les services bancaires et gère le système GAB.
- **Employé de banque** : Un individu responsable de la maintenance du GAB, y compris le rechargement en billets.

### 2. Identification des Cas d'Utilisation

#### Cas d'Utilisation Principal : Retrait d'argent
- **Acteur** : Client
- **Description** : Le client souhaite retirer de l'argent de son compte bancaire via le GAB.

#### Cas d'Utilisation secondaire 1 : Consultation du solde du compte
- **Acteur** : Client
- **Description** : Le client souhaite consulter le solde de son compte bancaire via le GAB.

#### Cas d'Utilisation secondaire 2 : Dépôt d'espèces
- **Acteur** : Client
- **Description** : Le client souhaite déposer de l'argent liquide sur son compte bancaire via le GAB.

#### Cas d'Utilisation secondaire 3 : Dépôt de chèques
- **Acteur** : Client
- **Description** : Le client souhaite déposer des chèques sur son compte bancaire via le GAB.

#### Cas d'Utilisation secondaire 4 : Recharge du GAB
- **Acteur** : Employé de banque
- **Description** : L'employé de banque recharge le GAB avec des billets de banque.

## Solution -> Exercice 2

### 1. Cas d'utilisation "Acheter un produit" et cas d'utilisation "Vérifier la disponibilité du produit"
- **Relation** : Association
- **Justification** :
  - Le cas d'utilisation "Vérifier la disponibilité du produit" fournit des informations nécessaires au cas d'utilisation "Acheter un produit".
  - Un client peut vouloir vérifier la disponibilité d'un produit avant de l'acheter.
  - Le cas d'utilisation "Acheter un produit" peut dépendre de la disponibilité du produit vérifiée dans le cas d'utilisation "Vérifier la disponibilité du produit".

### 2. Classe "Ordinateur" et classe "Système d'Exploitation"
- **Relation** : Composition
- **Justification** :
  - Un ordinateur ne peut pas fonctionner sans système d'exploitation.
  - Le système d'exploitation est une partie intégrante de l'ordinateur et gère ses ressources matérielles et logicielles.
  - La composition indique qu'un ordinateur existe nécessairement avec un système d'exploitation spécifique.

### 3. Classe "Outil" et classe "Marteau"
- **Relation** : Héritage
- **Justification** :
  - Un marteau est un type d'outil spécifique.
  - Un marteau hérite des caractéristiques et des fonctionnalités générales d'un outil.
  - Le marteau peut avoir des propriétés et des méthodes spécifiques qui le distinguent des autres outils.

### 4. Acteur "Peintre", acteur "Artiste" et acteur "Chanteur"
- **Relation** : Généralisation
- **Justification** :
  - Un peintre, un artiste et un chanteur sont tous des types d'artistes.
  - Ils partagent des caractéristiques et des compétences artistiques communes.
  - La généralisation indique qu'un artiste peut être un peintre, un chanteur ou un autre type d'artiste.

### 5. Cas d'utilisation "Jouer à la loterie" et cas d'utilisation "Gagner à la loterie"
- **Relation** : Association
- **Justification** :
  - Le cas d'utilisation "Gagner à la loterie" est un résultat possible du cas d'utilisation "Jouer à la loterie".
  - Un joueur ne peut gagner à la loterie qu'en y jouant d'abord.
  - Le cas d'utilisation "Gagner à la loterie" dépend du succès du cas d'utilisation "Jouer à la loterie".

### 6. Classe "Document" et classe "Feuille"
- **Relation** : Agrégation
- **Justification** :
  - Une feuille peut faire partie d'un document, mais elle peut également exister indépendamment.
  - Un document peut contenir plusieurs feuilles, mais il n'est pas nécessairement composé uniquement de feuilles.
  - L'agrégation indique qu'une feuille a une relation "est-une-partie-de" avec un document, mais elle peut aussi exister de manière autonome.
